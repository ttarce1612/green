Xét thí nghiệm gồm 2 lần phép thử Bernoulli có p = 0.4
1. Tính xác suất 1 lần thành công
2. Tính xác suất không lần nào thành công
3. Tính xác suất 2 lần thành công
4. Tính xác suất tối thiểu 1 lần thành công
5. Tính kỳ vọng, phương sai, độ lệch chuẩn
---

```python
## Thư viện
import math
import matplotlib.pyplot as plt
import numpy             as np
import pandas            as pd
import seaborn           as sns
import warnings
warnings.filterwarnings('ignore')

from scipy.stats         import binom, bernoulli
```

```python
## Tham số của mô hình
n = 2
p = 0.4
```

```python
## 1. Tính xác suất 1 lần thành công: P(X = 1) = f(1)
x     = 1
prob1 = binom.pmf(x, n, p)
print(prob1)
```

```python
## Tính toán theo công thức
def combination(n, x):
    comb  = math.factorial(n) /(math.factorial(x) * math.factorial(n - x))
    return comb
```

```python
prob2 = combination(n, x) * (p**x) * ((1 - p)**(n - x))
print(prob2)
```

```python
## 2. Tính xác suất không có lần nào thành công: P(X = 0) = f(0)
x     = 0
prob1 = binom.pmf(x, n, p)
print(prob1)

# Tính toán theo công thức
prob2 = combination(n, x) * (p**x) * ((1 - p)**(n - x))

print(prob2)
```

```python
## 3. Tính xác suất 2 lần thành công: f(2)
x     = 2
prob1 = binom.pmf(x, n, p)
print(prob1)

# Tính toán theo công thức
prob2 = combination(n, x) * (p**x) * ((1 - p)**(n - x))

print(prob2)
```

```python
## 4. Tính xác suất tối thiểu 1 lần thành công: P(X >= 1) = f(1) + f(2) = 1 - f(0)
print(1 - binom.pmf(0, n, p))
print(binom.pmf(1, n, p) + binom.pmf(2, n, p))
```

```python
## 5. Tính kỳ vọng, phương sai, độ lệch chuẩn
##    Hàm binom.stats(n, p) trả về expectation và variance
result = binom.stats(n, p)
print('(expectation, variance, standard deviation) =', result[0], result[1], math.sqrt(result[1]))

## Sử dụng công thức
print('(expectation, variance, standard deviation) =', n * p, n * p * (1 - p), math.sqrt(n * p * (1 - p)))
```

## **Bài toán 6:**
Xét một nhóm người có khoảng 20% mắc bệnh cao huyết áp.
Tiến hành lấy mẫu 1000 lần, mỗi lần lấy mẫu ngẫu nhiên là 20 người.
1. Giả lập số liệu theo phân phối nhị thức.
2. Vẽ biểu đồ histogram và biểu đồ KDE. Nêu nhận xét.
3. Tính các giá trị trung bình, trung vị, phương sai và độ lệch chuẩn.
4. Gọi X là số bệnh nhân trong mỗi mẫu. Tính các xác suất theo sự biến thiên của X từ 0 đến 20.
5. Vẽ biểu đồ bar chart thể hiện kết quả câu 4.
---

```python
## Các tham số của mô hình (Binomial Distribution)
trials = 1000
n      = 20
p      = 0.20
```

```python
## 1. Tạo số liệu ngẫu nhiên
rvs = binom.rvs(n = n, p = p, size = trials, random_state = 1)

## Quan sát dữ liệu
print(rvs[:50]) # số bệnh nhân trong mỗi lần lấy mẫu

ser = pd.Series(rvs)
ser.value_counts()
```

```python
## 2. Vẽ biểu đồ histogram và biểu đồ KDE. Nêu nhận xét.
sns.histplot(data = rvs, kde = True)
plt.show()
## Nhận xét: biểu đồ có dạng hình chuông
```

```python
## 3. Tính các giá trị trung bình, trung vị, phương sai và độ lệch chuẩn
print('(mean, median, var, std) =', np.mean(rvs), np.median(rvs), rvs.var(), rvs.std())
```

```python
## 4. Tính các xác suất theo sự biến thiên của X từ 0 đến 20.
for x in range(10):
    print('P( X =', x, ') =', str(round(binom.pmf(x, n, p),4) * 100) + '%')
```

```python
## 5. Vẽ biểu đồ bar chart thể hiện kết quả câu 4.
x = np.arange(10)
y = binom.pmf(x, n, p)
sns.barplot(x = x, y = y)
sns.lineplot(x = x, y = y)
plt.show()
```
