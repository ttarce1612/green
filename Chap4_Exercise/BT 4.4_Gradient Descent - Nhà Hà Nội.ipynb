## **Yêu cầu : Hồi quy tuyến tính bằng hàm của sklearn**
- Đọc dữ liệu vào dataframe
- Chọn input X = df[['Diện tích (m2)', 'Số phòng', 'Tuổi căn nhà', 'Số bếp lửa']]
- Chọn y = df['Giá (nghìn đôla)']
- Sử dụng linear_model.LinearRegression()
- Biểu diễn trực quan dữ liệu
- Dự đoán giá nhà x = (159, 7, 37, 0)
- Dự đoán giá nhà x = (205, 10, 20, 1)
- Dự đoán giá nhà x = (110, 5, 50, 0)
---

```python
## Thư viện
import numpy    as np
import pandas   as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn    import linear_model
```

```python
## Dữ liệu về nhà ở Hà Nội
df = pd.DataFrame([[100, 7, 42, 0],
                   [150, 7, 62, 0],
                   [118, 6, 40, 0],
                   [123, 6, 54, 0],
                   [112, 6, 42, 0],
                   [99, 6, 56, 0],
                   [124, 7, 51, 1],
                   [150, 6, 32, 0],
                   [342, 10, 42, 1],
                   [300, 9, 14, 1],
                   [123, 6, 32, 0],
                   [155, 6, 30, 0],
                   [98, 5, 30, 0],
                   [112, 6, 32, 0],
                   [102, 5, 46, 1],
                   [150, 6, 32, 0],
                   [166, 8, 50, 0],
                   [149, 7, 22, 1],
                   [138, 6, 17, 0],
                   [150, 7, 23, 0]
                  ],
                  columns = ['Diện tích (m2)', 'Số phòng', 'Tuổi căn nhà', 'Số bếp lửa'])

## Bổ sung thông tin về giá cả
df['Giá (nghìn đôla)'] = [259, 295, 279, 259, 299, 299, 309, 289, 849, 829, 359, 315, 310, 309, 300, 289, 369, 419, 405, 439]
```

```python
## Hiển thị dữ liệu
print(df.head())
```

```python
## Mô hình hóa dữ liệu: Input X, Output y
X = df[['Diện tích (m2)', 'Số phòng', 'Tuổi căn nhà', 'Số bếp lửa']]
y = df['Giá (nghìn đôla)']

## Hồi quy tuyến tính
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
## Các tham số: w0, w1, ..., wn
print('Intercept - Bias (w0):', regr.intercept_)
print('w1, w2, w3, w4)      :', regr.coef_) # slope
```

```python
print("Dự đoán giá nhà 159 mét vuông, 7 phòng, xây được 37 năm, không có bếp lửa: $",
      int(round(regr.predict([[159, 7, 37, 0]])[0],0)),"000")
print("Dự đoán giá nhà 205 mét vuông, 10 phòng, xây được 20 năm, có 1 bếp lửa: $",
      int(round(regr.predict([[205, 10, 20, 1]])[0],0)),"000")
print("Dự đoán giá nhà 110 mét vuông, 5 phòng, xây được 50 năm, không có bếp lửa: $",
      int(round(regr.predict([[110, 5, 50, 0]])[0],0)),"000")
```

```python
## Chọn thuộc tính Input dựa trên mức độ tương quan (Pearson correlation)
df.corr()
## --> Chọn 'Diện tích (m2)' vì có hệ số tương quan cao nhất đối với 'Giá' (0.941754)
```
