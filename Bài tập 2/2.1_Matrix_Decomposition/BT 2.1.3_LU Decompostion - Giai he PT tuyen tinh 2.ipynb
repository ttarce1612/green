```python
## Thư viện
import numpy  as np
import scipy
import warnings
warnings.filterwarnings('ignore')
import scipy.linalg as linalg
from scipy.linalg import lu, inv
```

```python
## Ma trận hệ số A và vector hệ số tự do B
A = np.array([[1, 2, 4],[3, 8, 14],[2, 6, 13]])
B = np.array([3, 13, 4])
print(A)
print(B)
```

```python
## Tìm nghiệm bằng ma trận nghịch đảo
X = inv(A) @ B
X
```

```python
## Áp dụng phân rã LU
P, L, U = lu(A)
```

```python
## Kiểm chứng phép phân rã
(P @ L @ U).astype(int)
```

```python
## B1. Giải L.Y = P.T @ B
Y = inv(L) @ P.T @ B
Y
```

```python
## B2. Giải U.X = Y
X = inv(U) @ Y
X
```

```python
## B2. Giải U.X = Y
X = inv(U) @ inv(L) @ P.T @ B
X
```
