```python
## Thư viện
import numpy    as np
import pandas   as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
## Mở tập tin dữ liệu iris.csv
iris = pd.read_csv(folder + '/Data/iris.csv')

## Xem dữ liệu
print(iris.shape)
print(iris.head())
```

```python
## Chỉ sử dụng 4 cột đầu tiên (cột 'species' là nhãn phân loại - target)
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
X.head()
```

```python
## Phân tích SVD
U, s, VT = np.linalg.svd(X)
```

```python
## Hiển thị kết quả
print('* Ma trận U(', U.shape[1], ', ', U.shape[0], '): \n', U, '\n')
print('* Singular values:', s, '\n')
print('* Ma trận VT(', VT.shape[1], ', ', VT.shape[0], '): \n', VT, '\n')
```

### **Nhận xét:**
Hai trị riêng đầu tiên có cường độ vượt trội so với các trị riêng còn lại. Điều đó có nghĩa là 2 vectors riêng đầu tiên đóng vai trò quan trọng hơn rất nhiều và chúng ta có thể tái tạo dữ liệu chỉ bằng hai hàng đầu tiên của V và hai cột đầu tiên của U.

```python
## Chỉ sử dụng 2 singular values và 2 eigen vectors đầu tiên
D = np.matrix(U[:, :2]) * np.diag(s[:2]) * np.matrix(VT[:2, :])
print('- Ma trận D(', D.shape[0], ',', D.shape[1], '): \n', D[0:5], '\n')
```

```python
## Chuyển vào dataframe
Xnew = pd.DataFrame(D, index = X.index, columns = X.columns)
print(Xnew.head())
```

```python
## Sai số so với dữ liệu ban đầu: (X - D)
print((X - D).head().to_string(index = False))
```

```python
## Tái tạo ma trận dữ liệu (giữ nguyên các thành phần)
Sigma = np.zeros((X.shape[0], X.shape[1]))
Sigma[:X.shape[1], :X.shape[1]] = np.diag(s)

X_c = U @ Sigma @ VT
X_c[0:5]
```

```python
## Chuyển vào dataframe
X_c_new = pd.DataFrame(X_c, columns  =X.columns)
X_c_new.head()
```
