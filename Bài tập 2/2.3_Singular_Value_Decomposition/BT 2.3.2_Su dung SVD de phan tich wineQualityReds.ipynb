```python
## Thư viện
import numpy      as np
import pandas     as pd
import warnings
warnings.filterwarnings('ignore')

from scipy.linalg import svd
```

```python
## Mở tập tin dữ liệu
data = pd.read_csv(folder + '/Data/wineQualityReds.csv', index_col = 0)

## Kích thước dữ liệu
data.shape # 1599 dòng và 12 cột
```

```python
## Xem những dòng dữ liệu đầu tiên
data.head()
```

```python
## Xem những dòng dữ liệu cuối cùng
data.tail()
```

```python
## Áp dụng phân rã SVD
U, s, V_T = svd(data)
```

```python
## Ma trận U[m, m] (left-eigen vectors)
print('Ma trận U' , U.shape, ':\n', U)
```

```python
## Singular values
s.tolist()
```

```python
## Tạo ma trận S[m, n]
Sigma = np.zeros(data.shape)
Sigma[:s.size, :s.size] = np.diag(s)
Sigma.shape
```

```python
## Ma trận V.T[n, n] (right-eigen vectors)
print('\nMa trận V.T', V_T.shape, ':\n', V_T)
```

```python
## Tái tạo dữ liệu ban đầu
print('\n* Kiểm chứng lại phép phân rã:\n', U @ Sigma @ V_T)
```
