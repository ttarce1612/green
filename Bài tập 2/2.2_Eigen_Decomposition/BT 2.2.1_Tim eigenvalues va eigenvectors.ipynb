## **Yêu cầu 1:**
- Tạo ma trận A(4x4) với các giá trị ngẫu nhiên từ 1 đến 10
- Tìm eigenvalues (lambda[i]) và eigenvectors (v[i]) của A
- Kiểm chứng eigenvector đầu tiên: A.v[0] = lambda[0].v[0] ?
- Nếu bằng nhau thì tái tạo A từ các eigenvectors và eigenvalues
---

```python
## Thư viện
import numpy    as np
import warnings
warnings.filterwarnings('ignore')

from numpy        import diag
from numpy.linalg import eig, inv
```

```python
## Tạo ma trận A(4, 4) với các giá trị ngẫu nhiên từ 1 đến 10
m, n, min, max = 4, 4, 1, 10
A = np.random.randint(min, max + 1, (m, n))
A
```

```python
## Tìm eigenvalues và eigenvectors của A
values, vectors = eig(A)
```

```python
## Hiển thị kết quả: Trị riêng lambda(r: số thực, i: số phức)
for j in range(len(values)):
    eigenvalue = values[j]

    r = values[j].real
    i = values[j].imag
    print('Eigenvalue[%1d]' %j, '= (%f,' %r, '%f)' %i)

print('\nEigenvectors', vectors.shape, ':\n', vectors)
```

```python
## Kiểm chứng eigenvector đầu tiên: A.v[0] = lambda[0].v[0] ?
v0      = vectors[:, 0]
lambda0 = values[0]     # Chú ý phần số thực
print('A @ v[0]         =', A @ v0)
print('lambda[0] @ v[0] =', lambda0 * v0)
```

```python
## Tái tạo A từ các eigenvectors và eigenvalues:A = Q @ Lambda @ Q_1
Q = vectors
Lambda = diag(values)

print('\nTái tạo ma trận A:\n', np.rint(Q @ Lambda @ inv(Q)).astype(int))
# Nhận xét: A có phải là ma trận chéo hóa được ?
```

```python
## Hàm eigvalsh() chỉ trả về eigenvalues (khôngtạo eigenvectors)
eigenvalues = np.linalg.eigvalsh(A)
print(eigenvalues)
```

       [ 9, 11,  1,  2]
       [ 7,  5, 10,  4]

```python
A = np.array([[ 9, 11,  1,  2],
              [ 7,  5, 10,  4],
              [ 2, 10,  1,  1],
              [ 1,  1, 10,  9]])

## Tìm eigenvalues và eigenvectors của A
values, vectors = eig(A)
print('Eigenvalues :', values)
print('Eigenvectors:\n', vectors)
```

```python
## Tái tạo A từ các eigenvectors và eigenvalues:A = Q @ Lambda @ Q_1
Q = vectors
Lambda = diag(values)
print('\nTái tạo ma trận A:\n', np.rint(Q @ Lambda @ inv(Q)).astype(int))
```
