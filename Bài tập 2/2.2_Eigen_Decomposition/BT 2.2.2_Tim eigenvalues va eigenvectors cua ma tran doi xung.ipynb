```python
import numpy      as np
import warnings
warnings.filterwarnings('ignore')

from numpy.linalg import eig, inv
from numpy        import diag
```

```python
## Tạo ma trận A(5 x 5) chứa các giá trị ngẫu nhiên trong khoảng 1 - 10
m, n, min, max = 5, 5, 1, 10
A = np.random.randint(min, max + 1, (m, n))
print('Ma trận A', A.shape, ':\n', A)
```

```python
## Tạo ma trận đối xứng  B = A @ A.T
B = A @ A.T
B
```

```python
## Tìm eigenvalues và eigenvectors của B
values, vectors = eig(B)
```

```python
## Hiển thị kết quả: Trị riêng lambda(r: số thực, i: số phức)
for j in range(len(values)):
    eigenvalue = values[j]

    r = values[j].real
    i = values[j].imag
    print('Eigenvalue[%1d]' %j, '= (%f,' %r, '%f)' %i)

print('\nEigenvectors', vectors.shape, ':\n', vectors)
```

---
### **Nhận xét:**
- Ma trận ĐỐI XỨNG có các giá trị riêng đều là SỐ THỰC !
---

```python
## Tái tạo B từ các eigenvectors và eigenvalues: B = Q @ Lambda @ Q_1
from numpy.linalg import inv
Q = vectors
Lambda = diag(values)

print('Tái tạo ma trận B:\n', np.rint(Q @ Lambda @ inv(Q)).astype(int))
```
