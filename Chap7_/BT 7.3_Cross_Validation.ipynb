```python
## 1. Đọc tập tin dữ liệu vào dataframe
data = pd.read_csv(folder + "/Data/fruit_data_with_colors.txt", sep = '\t')

## 2. In các thông tin: shape, info, head.
print(data.shape)
print(data.info())
print(data.head())
```

```python
## 3. Loại bỏ 2 cột fruit_name và fruit_subtype khỏi dataframe.
##    Chọn y là cột fruit_label và X gồm các cột còn lại.
X = data[['mass', 'width', 'height', 'color_score']]
y = data['fruit_label']
```

```python
## 3. Áp dụng train/test split cho dữ liệu X, y theo tỷ lệ 70:30.
##    In các index của mỗi bộ.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

print(X_train.index)
print(X_test.index)
```

```python
## 4. Áp dụng K-fold với k = 10
cv = KFold(n_splits = 10, shuffle = True)

## In các index của mỗi bộ.
k = 1
for train_index, test_index in cv.split(X):
    print(f"** k = {k}")
    print("  - Train Index:", train_index.tolist())
    print("  - Test Index: ", test_index.tolist(), "\n")
    X_train, X_test = X.iloc[train_index.tolist()], X.iloc[test_index.tolist()]
    y_train, y_test = y.iloc[train_index.tolist()], y.iloc[test_index.tolist()]

    k += 1
```
