```python
##------------------------------------------------------------------------------
## Hàm tính bách phân vị thứ p của dãy số
##    index = (p * n) / 100
##------------------------------------------------------------------------------
def percentile(data, p):

    data = sorted(data)
    size = len(data)

    if (p == 0):
        value = data[0]
    elif (p == 100):
        value = data[size - 1]
    else:
        # Tính chỉ số index
        i = (p * size) / 100
        k = int(i)

        # Tính phân vị thứ p
        if (k == i):
            print('i =', i, 'k =', k)

            value = (data[k - 1] + data[k]) / 2
        else:
            value = data[k]

        ## Interpolation
        ## f = i - k
        ## value = ((1 - f) * x[k - 1]) + (f * x[k])
    return value
```

```python
##------------------------------------------------------------------------------
## Hàm hiển thị kết quả hàm percentile() với các giá trị tham số 'method'
##------------------------------------------------------------------------------
def percentile_python(x, p):
    x = sorted(x)
    percentiles    = np.zeros(5)
    percentiles[0] = np.percentile(x, p) # Default: linear
    percentiles[1] = np.percentile(x, p, method = 'lower')
    percentiles[2] = np.percentile(x, p, method = 'higher')
    percentiles[3] = np.percentile(x, p, method = 'midpoint')
    percentiles[4] = np.percentile(x, p, method = 'nearest')

    return percentiles
##------------------------------------------------------------------------------
```

```python
## Dãy số
x = np.array([8, 9, 12, 14.5, 15.5, 17, 18])

## Các giá trị thử nghiệm của p ([0..100])
p = list(range(0, 105, 5))
```

```python
print('** Sử dụng hàm np.percentile()')
##------------------------------------------------------------------------------
## METHOD parameter --> khi index = 1 + (p * (n - 1)) / 100 không là số nguyên,
##    index thuộc khoảng (L, U)
##    ‘linear’: x[L] + (x[U] - x[L]) * f; (f là phần thập phân của index)
##    ‘lower’: x[L]
##    ‘higher’: x[U]
##    ‘midpoint’: (x[L] + x[U]) / 2
##    ‘nearest’: làm tròn index
##------------------------------------------------------------------------------
percentiles = np.zeros((len(p), 6))

for i in range(0, len(p)):
    print('   p%d =' %p[i], percentile_python(x, p[i]))
```

```python
print('** Sử dụng công thức')
for i in range(0, len(p)):
    print('   p%d =' %p[i], percentile(x, p[i]))
```
