## Câu 1. Giảm chiều dữ liệu (2 điểm)
Tập tin ‘Breast-Cancer.csv’ chứa dữ liệu phân lớp bệnh nhân ung thư (Classification  {1, 2}) dựa trên các thuộc tính: Age, BMI, Glucose, Insulin, HOMA, Leptin, Adiponectin, Resistin và MCP.1.

 1.1) Áp dụng phương pháp PCA để giảm xuống còn k chiều (k > 2) so với dữ liệu gốc. Giải thích nguyên nhân (hay cơ sở) của số chiều được giảm.
 
 1.2) Giảm chiều xuống còn k = 2 và trực quan hóa dữ liệu. Nhận xét kết quả.


```python
## Thư viện
import matplotlib.pyplot   as plt
import numpy               as np
import pandas              as pd
import seaborn             as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.decomposition import PCA
```

```python
## Đọc tập tin vào dataframe
df     = pd.read_csv('Data/Breast-Cancer.csv')
#print('df =', df)

target = 'Classification'
print('* Biến phân lớp:', target)

## Danh sách các features
nb_features = df.shape[1] - 1
features_names    = df.columns[:nb_features]

print('* Số lượng features = %2d' %nb_features)
print('* Các features:', ', '.join(features_names), '\n')

print(df.head())
```

```python
## Áp dụng PCA để giảm xuống k chiều (k > 2)
from sklearn.decomposition import PCA

## Loại bỏ cột Class trong dataframe
df_without_target = df.drop(columns = ['Classification'])

#   https://stackoverflow.com/questions/32857029/python-scikit-learn-pca-explained-variance-ratio-cutoff
#   - The pca.explained_variance_ratio_ returns a vector of the variance explained by each dimension.
#   - The pca.explained_variance_ratio_[i] gives the variance explained solely by the i+1st dimension.
#   - The pca.explained_variance_ratio_.cumsum() will return a vector x
#     such that x[i] returns the cumulative variance explained by the first i+1 dimensions.

#   (1) PCA().components_: Chuyển vị của ma trận vectơ riêng EigenVectors.T
#   (2) PCA().explained_variance_: Các giá trị riêng
#   (3) PCA().explained_variance_ratio_: Tỷ lệ phương sai so với dữ liệu gốc
#   (4) Hàm numpy.cumsum()
df_without_target.head()
```

### Câu 1.1:  Áp dụng phương pháp PCA để giảm xuống còn k chiều (k > 2) so với dữ liệu gốc. Giải thích nguyên nhân (hay cơ sở) của số chiều được giảm.


```python
## Áp dụng PCA với k CHƯA được xác định
##    số chiều ngầm định:  n_components = min(n_samples, n_features)
pca = PCA().fit(df_without_target) # PCA().fit(): thực hiện PCA
pca
```

```python
## Phương sai trên từng chiều
print(pca.explained_variance_.flatten())
```

```python
# Các điểm dữ liệu

points = np.cumsum(pca.explained_variance_ratio_) * 100
points = np.insert(points, 0, 0) # Thêm điểm k = 0, variance = 0
print('points:', points)

x_i = np.arange(0, nb_features + 1)
print('x_i:', x_i)
y_i = (points[0:])//0.01/100
y_i2 = (points[-10:])//0.01/100
print('y_i1:', y_i)
print('y_i2:', y_i)

plt.figure(figsize = (8, 6))
plt.plot(points, marker = 'o')
plt.xlabel('Số features (k)')
plt.ylabel('Variance (%)')
plt.title('Đồ thị biểu diễn % phương sai tích lũy theo số features (k)')
plt.xlim([0, nb_features + 1])
plt.grid(axis = 'x')
for i in x_i:
    plt.text(i, y_i[i] + 1, round(y_i[i], 1), ha = 'center', va = 'baseline')
    # tung độ của text cao hơn point 1 đơn vị

plt.show()
```

```python
## Kiểm chứng: Tính phương sai tích lũy theo giá trị của k
var = 0.0
for k in range(1, nb_features + 1):
    pca = PCA(k)
    pca.fit(df_without_target)

    newVar = pca.explained_variance_ratio_.sum() * 100
    print('   * k = %2d' %k, ': phương sai tích lũy ~ %.1f%%' %newVar,
          '--> tăng ~ %.1f%%' %(newVar - var))
    var = newVar
```

#### Gỉai thích: chọn k = 3 vì đã giữ lại dược 99.6% dữ liệu 


### Câu 1.2 Giảm chiều xuống còn k = 2 và trực quan hóa dữ liệu. Nhận xét kết quả.

```python
# Áp dụng PCA giảm xuống còn k = 2 chiều so với dữ liệu gốc

# A = df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']].values

k = 2
pca = PCA(n_components = k)
pca.fit(df_without_target)
```

```python
## Phương sai trên từng chiều
# print(pca.explained_variance_.flatten())
```

```python
points = np.cumsum(pca.explained_variance_ratio_) * 100
points
```

```python
## Hiển thị kết quả tóm tắt
#  components_: array, shape (n_components, n_features)
#               Các thành phần chính trong không gian features,
#               biểu thị các hướng của phương sai tối đa trong dữ liệu
#  explained_variance_ : array, shape (n_components,)
#               Lượng phương sai được giải thích bởi từng thành phần được chọn.
print('PCA.components:\n', pca.components_)
print('PCA.shape:', pca.components_.shape)
print('PCA.explained variance:', pca.explained_variance_)
print('PCA.explained variance shape:', pca.explained_variance_.shape)
```

```python
## Phép chiếu dữ liệu vào không gian MỚI (tọa độ mới)

P = pca.transform(df_without_target)
P[0:5]
```

```python
## Biểu diễn trực quan dữ liệu
plt.figure(figsize = (6, 6))
plt.scatter(P[:, 0], P[:, 1])
plt.show()
```

```python
## Gán column names cho các cột
df_pca = pd.DataFrame(P, columns = ['PC 1', 'PC 2'])
df_pca.head()
```

```python
## Ghép nối thêm cột phân lớp (target = 'classification'), đặt tên là cột 'classification'
df_pca['classification'] = df['Classification']
df_pca.head()
```

```python
## Biểu diễn trực quan dữ liệu sau khi áp dụng PCA
plt.figure(figsize = (6, 6))
sns.scatterplot(x = "PC 1", y = "PC 2", data = df_pca, hue = 'classification')
plt.show()
```

```python
## Biểu diễn trực quan dữ liệu BAN ĐẦU
plt.figure(figsize = (6, 6))
sns.scatterplot(x = "Age", y = "BMI", data = df, hue = 'Classification')
plt.show()
```
